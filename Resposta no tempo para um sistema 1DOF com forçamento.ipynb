```python
# Esta célula importa os pacotes que iremos utilizar
import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
# O comando abaixo é utilizado para que a figura seja mostrada no próprio notebook
%matplotlib inline 
```

```python
#Condições Inicias
v0 = 0
x0 = 0.1
#Parametros Dados e Passo
wn = 2*2*sp.pi
wf = 2.1*2*sp.pi
f0 = 1
t = sp.linspace(0,50,5000)
```

```python
def f(wf): # Resposta no tempo para um sistema 1DOF com forçamento
    x = (v0/wn)*sp.sin(wn*t) + (x0 - (f0/(wn**2 - wf**2)))*sp.cos(wn*t) + (f0/(wn**2 - wf**2))*sp.cos(wf*t)
    plt.plot(t,x)
```

```python
interact(f, wf=1.5)
```
